```python
import numpy as np
import pandas as pd
```

```python
import pyarrow.parquet as pq
trips = pq.read_table('yellow_tripdata_2024-01.parquet')
trips = trips.to_pandas()
print(trips.shape)
trips.head()
```

```python
trips['tpep_pickup_datetime'] = pd.to_datetime(trips['tpep_pickup_datetime'])
trips['tpep_dropoff_datetime'] = pd.to_datetime(trips['tpep_dropoff_datetime'])
trips['duration'] = trips['tpep_dropoff_datetime'] - trips['tpep_pickup_datetime']
trips['duration'] = trips.duration.apply(lambda x: x.total_seconds()/60)
```

```python
trips.describe()
```

```python
trips.head()
```

```python
filtered_df = trips[(trips['duration'] >=1) & (trips['duration'] <=60)]
filtered_df.shape
```

```python
(filtered_df.shape[0]) / trips.shape[0]
```

```python
categorical_data = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
```

```python
filtered_df[categorical_data] = filtered_df[categorical_data].astype(str)
```

```python
df = filtered_df[categorical_data + numerical]
df
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dicts = df.to_dict(orient = 'records')
```

```python
dv = DictVectorizer()
x_train = dv.fit_transform(train_dicts)
```

```python
len(dv.feature_names_)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
y_train = filtered_df.duration
```

```python
from sklearn.metrics import mean_squared_error
```

```python
lr.fit(x_train, y_train)
```

```python
y_predict = lr.predict(x_train)
```

```python
rmse = np.sqrt(mean_squared_error(y_train, y_predict))
```
